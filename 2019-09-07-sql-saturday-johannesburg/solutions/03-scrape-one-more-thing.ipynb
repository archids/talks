Let's see if we can scrape some data from a local law firm. All of the data are on a single page so, in principle, this should be an easy scrape!

## Setup

Load some libraries.

```r
suppressMessages(library(rvest))
```

The URL. Open [this link](https://www.bowmanslaw.com/our-people/) in your browser.

The first hint that something is different is the distinct delay in page load. What's happening there?

```r
URL = "https://www.bowmanslaw.com/our-people/"
```

## Scraping

The first step will be to grab the HTML for the page.

```r
html <- read_html(URL)
```

Now we need to get a selector. We'll start by targeting the lawyers' names. Using the browser you'll find that the required selector is `.media-heading`.

```r
html %>% html_nodes(".media-heading")
```

No results? What's happening here?

Here's the problem: the content on this page is being generated by JavaScript. So it's happening on the client (the browsers) not the server. So the *static* web scraping tools that we've been looking at will no longer work.

There's a different set of tools for scraping *dynamic* web sites. But that's another story for another day.
